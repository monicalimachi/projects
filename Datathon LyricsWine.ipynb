```python
# Libraries Required
import pandas as pd
import nltk
nltk.download('all')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

```

```python
# Load dataset
train = pd.read_csv('datathonwine/billboard_lyrics_1964-2015.csv', encoding="latin-1")
```

```python
train.head()
```

```python
df = train.sort_values(['Artist'],ascending=[1])
train.sort_values(by='Artist')
Counter(df["Artist"])
```

```python
artist_song = train.query ("Artist == 'ariana grande and the weeknd'")
print(artist_song["Song"])
```

```python
artist_song = train.query ("Artist == 'ariana grande and the weeknd' and Song == 'love me harder'")
Lyrics=str(artist_song['Lyrics'].values)
print(Lyrics)
```

```python
#words = [word for word in words if word not in all_stopwords]

stop_words = set(stopwords.words('english'))
 
word_tokens = word_tokenize(Lyrics)
 
filtered_sentence = [w for w in word_tokens if not w in stop_words]
 
filtered_sentence = []
 
for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
 
#print(stop_words)
print(filtered_sentence)
```

```python
lyrics_vec=Lyrics.split( " " )
```

```python
fdist = nltk.FreqDist(lyrics_vec)

for word, frequency in fdist.most_common():
    print('{};{}'.format(word, frequency))
```

```python
# Calculate frequency distribution
fdist = nltk.FreqDist(filtered_sentence)

for word, frequency in fdist.most_common():
    print('{};{}'.format(word, frequency))
```
